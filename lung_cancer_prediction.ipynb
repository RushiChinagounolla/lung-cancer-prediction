```python
# Defining a function to categorize chronic lung disease severity based on level
def CLD_Severity(x):
    if x <= 2:
        return 'Mild'
    elif x <= 5:
        return 'Moderate'
    else:
        return 'Severe'
# Creating a new column for severity classification by applying the function to the 'chronic Lung Disease' column
df['CLD_Severity'] = df['chronic Lung Disease'].apply(CLD_Severity)

# Visualizing the relation between smoking and CLD severity using a countplot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
# Creating a countplot to show the number of patients at each smoking level, separated by CLD severity
sns.countplot(data=df, x='Smoking', hue='CLD_Severity', palette=reversed(sns.color_palette("Set1", n_colors=3)))


plt.xlabel('Smoking Level')
plt.ylabel('Number of Patients')
plt.title('Smoking vs. Chronic Lung Disease Severity')
plt.legend(title='CLD Severity')
plt.tight_layout()
plt.show()
```

### Continuing analysis and preparing data for machine learning


```python
# Checking the data types of columns with 'object' type
df.dtypes[df.dtypes == 'object']
```

```python
# Importing LabelEncoder to encode categorical target labels into numeric form
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
# Mapping 'Cancer_Level' from string to numerical values
df["Cancer_Level"] = df["Cancer_Level"].map({'Low': 0, 'Medium': 1, 'High': 2})
# Dropping the 'CLD_Severity' column as it's not needed for modeling
df.drop(columns=['CLD_Severity'], inplace=True)

from sklearn.preprocessing import MinMaxScaler

# Define target and features
# Separating features (X_unscaled) from the target variable (y)
X_unscaled = df.drop("Cancer_Level", axis=1)
y = df["Cancer_Level"]

# Create scaled version for Logistic Regression
# Initializing the MinMaxScaler
scaler = MinMaxScaler()
# Scaling the features to a range between 0 and 1
X_scaled = scaler.fit_transform(X_unscaled)
```

```python
from sklearn.model_selection import train_test_split

# Splitting the unscaled data into training and testing sets
X_train_unscaled, X_test_unscaled, y_train, y_test = train_test_split(X_unscaled, y, test_size=0.3, random_state=0, stratify=y)
# Splitting the scaled data into training and testing sets
X_train_scaled, X_test_scaled, y_train,y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=0, stratify=y)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Define models to evaluate (excluding Logistic Regression here)
model_list = [
    DecisionTreeClassifier(random_state=0),
    RandomForestClassifier(n_estimators=100, random_state=0)
]

# Looping through each model in the list
for model in model_list:
    print(f"\n Training model: {model.__class__.__name__}")

    # Train the model using the unscaled training data
    model.fit(X_train_unscaled, y_train)

    # Making predictions on the test data
    test_pred = model.predict(X_test_unscaled)
    print("\n--- Test Results ---")
    # Displaying a confusion matrix for the test results
    display(pd.crosstab(y_test, test_pred, rownames=["Actual"], colnames=["Predicted"]))
    # Printing the accuracy score for the test data
    print("Test Accuracy:", round(accuracy_score(y_test, test_pred), 4))

    # Making predictions on the training data
    train_pred = model.predict(X_train_unscaled)
    print("\n=== Train Results ===")
    # Displaying a confusion matrix for the training results
    display(pd.crosstab(y_train, train_pred, rownames=["Actual"], colnames=["Predicted"]))
    # Printing the accuracy score for the training data
    print("Train Accuracy:", round(accuracy_score(y_train, train_pred), 4))

    print("=" * 50)
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize Logistic Regression model with a higher max_iter
log_model = LogisticRegression(random_state=0, max_iter=1000)

print(f"\n Training model: {log_model.__class__.__name__}")

# Train the Logistic Regression model using the scaled training data
log_model.fit(X_train_scaled, y_train)

# Making predictions on the scaled test data
y_pred_log_test = log_model.predict(X_test_scaled)
print("\n--- Test Results ---")
# Displaying a confusion matrix for the test results
display(pd.crosstab(y_test, y_pred_log_test, rownames=["Actual"], colnames=["Predicted"]))
# Printing the accuracy score for the test data
print("Test Accuracy:", round(accuracy_score(y_test, y_pred_log_test), 4))

# Making predictions on the scaled training data
y_pred_log_train = log_model.predict(X_train_scaled)
print("\n=== Train Results ===")
# Displaying a confusion matrix for the training results
display(pd.crosstab(y_train, y_pred_log_train, rownames=["Actual"], colnames=["Predicted"]))
# Printing the accuracy score for the training data
print("Train Accuracy:", round(accuracy_score(y_train, y_pred_log_train), 4))

print("=" * 50)
```

```python
from sklearn.metrics import classification_report

# Decision Tree Classification Report
print("=== Decision Tree Classification Report ===")
dt_preds = model_list[0].predict(X_test_unscaled)
print("Test Accuracy:", round(accuracy_score(y_test,dt_preds ), 4))
print(classification_report(y_test, dt_preds))

# Random Forest Classification Report
print("\n=== Random Forest Classification Report ===")
rf_preds = model_list[1].predict(X_test_unscaled)
print("Test Accuracy:", round(accuracy_score(y_test, rf_preds), 4))
print(classification_report(y_test, rf_preds))

# Logistic Regression Classification Report
print("\n=== Logistic Regression Classification Report ===")
lr_preds = log_model.predict(X_test_scaled)
print("Test Accuracy:", round(accuracy_score(y_test,lr_preds), 4))
print(classification_report(y_test, lr_preds))
```

```python
from sklearn.metrics import confusion_matrix

# Dictionary containing model names and their predictions
models = {
    "Decision Tree": model_list[0].predict(X_test_unscaled),
    "Random Forest": model_list[1].predict(X_test_unscaled),
    "Logistic Regression": log_model.predict(X_test_scaled)
}

# Looping through each model and its predictions to generate and display confusion matrices
for name, preds in models.items():
    # Calculating the confusion matrix
    cm = confusion_matrix(y_test, preds)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix: {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
```

```python
# Comparing the accuracy of different models using a bar plot
model_names = ["Decision Tree", "Random Forest", "Logistic Regression"]
# Calculating the accuracy scores for each model on the test data
accuracies = [
    accuracy_score(y_test, model_list[0].predict(X_test_unscaled)),
    accuracy_score(y_test, model_list[1].predict(X_test_unscaled)),
    accuracy_score(y_test, log_model.predict(X_test_scaled))
]

plt.figure(figsize=(8, 5))
# Creating a bar plot to compare the accuracies of the models
sns.barplot(x=model_names, y=accuracies, hue=model_names, palette='viridis', legend=False)
# Setting the y-axis limit
plt.ylim(0, 1)
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.show()
```
